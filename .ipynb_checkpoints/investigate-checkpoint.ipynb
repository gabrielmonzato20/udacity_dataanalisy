```python
data_gun= pd.read_excel('gun_data.xlsx')
data_gun.head()
```

```python
# Load your data and print out a few lines. Perform operations to inspect data
#   types and look for instances of missing or possibly errant data.
data_census = pd.read_csv('U.S. Census Data.csv')
data_census.head(5)
```

```python
data_gun.fillna(data_gun.mean(),axis=0,inplace=True)
#fist sum is for each column the second is for the hole dataset
data_gun.isnull().sum().sum()
```

```python
#this code is for drop all nun values 
data_census=data_census.drop(index=data_census.iloc[65:].index,axis=1)
```

```python
#i drop fact note because we has just nan volues and this dont help in analisty
data_census=data_census.drop('Fact Note',axis=1)
```

```python
data_census.fillna(data_census.mean(),axis=0,inplace=True)
```

```python
data_census.isnull().sum().sum()
```

```python
#now he dont have problems with null values i see some values in data_census dont was in right datatype so lets fix this
```

```python
#fist in data_census;covert by string to float  r'D' is a regular expression
```

```python
states=[x for x in data_census.columns[2:]]
for state in states:
    data_census[state].replace(regex=True,inplace=True,to_replace=r'\D',value=r'')
    data_census[state]=pd.to_numeric(data_census[state], downcast='float', errors='ignore')

```

```python
# change the columns  of data_gun for the rigth datatype
col= list(data_gun.columns)
col = col[2:]
data_gun[col] = data_gun[col].applymap(num.int64)
data_gun.info()


```

### Research Question 1 (Which states have had the highest growth in gun registrations?)

```python
#fist we nedd calculate the grot in the range og 2017_toals until 2016
dataframe['grothw'] = dataframe['2016_totals']-dataframe['2010_totals']
plot_total=dataframe.plot(kind='bar',x='state',y='grothw',figsize=(20,20))
plot_total.set_title("The increse of total guns from 2010 until 2016")
plot_total.set_xlabel("State")
plot_total.set_ylabel("Total")
```

### Research Question 2  (How state has the higthst number of  population estimated in 2010 )

```python
dataframe['Population estimates, July 1, 2016,  (V2016)'].values
dataframe['Population estimates, July 1, 2016,  (V2016)'][0]=dataframe['Population estimates, July 1, 2016,  (V2016)'][0].replace(",","")
dataframe['Population estimates, July 1, 2016,  (V2016)'][0]=dataframe['Population estimates, July 1, 2016,  (V2016)'][0].replace(".","")
```

```python
dataframe['Population estimates, July 1, 2016,  (V2016)']=dataframe['Population estimates, July 1, 2016,  (V2016)'].astype(str).astype(float)
```

```python
dataframe.columns
```

```python
dataframe['Population estimates base, April 1, 2010,  (V2016)'][0]=dataframe['Population estimates base, April 1, 2010,  (V2016)'][0].replace(",",'')
```

```python
dataframe['Population estimates base, April 1, 2010,  (V2016)']=dataframe['Population estimates base, April 1, 2010,  (V2016)'].astype(str).astype(float)
```

```python
dataframe['population_grothw'] = dataframe['Population estimates, July 1, 2016,  (V2016)']-dataframe['Population estimates base, April 1, 2010,  (V2016)']
```

```python
population_grotw=dataframe.plot(kind='bar',x='state',y='population_grothw',figsize=(20,20))
population_grotw.set_title("The increse of population growth per state (2010-2016)")
population_grotw.set_xlabel("state")
population_grotw.set_ylabel("Population Estimatived Grothw")
```

<h1>Question 3 (Whats the number of increse under 18 years people(2010-2016))</h1>

```python
dataframe['Persons under 18 years, percent, April 1, 2010'][0]=dataframe['Persons under 18 years, percent, April 1, 2010'][0].replace(".","").replace("%","")
dataframe['Persons under 18 years, percent, April 1, 2010'].values
```

```python
dataframe['Persons under 18 years, percent, July 1, 2016,  (V2016)']=dataframe['Persons under 18 years, percent, July 1, 2016,  (V2016)'].astype(str).astype(float)
```

```python
dataframe['Persons under 18 years, percent, April 1, 2010']=dataframe['Persons under 18 years, percent, April 1, 2010'].astype(str).astype(float)
```

```python
dataframe['under18_grothw']=dataframe['Persons under 18 years, percent, April 1, 2010']-dataframe['Persons under 18 years, percent, July 1, 2016,  (V2016)']
```

```python
plot_under_18=dataframe.plot(kind='bar',x='state',y='under18_grothw',title='The growth of under 18 peoples',figsize=(15,15))
plot_under_18.set_xlabel("state")
plot_under_18.set_ylabel("The growth total of under_18")
```
